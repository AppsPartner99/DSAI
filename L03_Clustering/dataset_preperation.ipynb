```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_correlated_dataset(n, covariance, mu, scale):
    np.random.seed(0)
    latent = np.random.randn(n, 2)
    dependent = latent.dot(covariance)
    scaled = dependent * scale
    scaled_with_offset = scaled + mu
    
    # return x and y of the new, correlated dataset
    return scaled_with_offset[:, 0], scaled_with_offset[:, 1]
```

```python
centroid_1 = [2,4]
centroid_2 = [14,12]
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

covariance = [[0.9, 0.5],
              [0.15, -0.35]]

x_1, y_1 = get_correlated_dataset(800, covariance, mu=centroid_1, scale=[3, 2])
x_2, y_2 = get_correlated_dataset(800, covariance, mu=centroid_2, scale=[2, 3])

ax.scatter(x_1, y_1, s=10, color='#79AEA3')
ax.scatter(x_2, y_2, s=10, color='#DB5461')

ax.scatter(centroid_1[0], centroid_1[1], s=50, label='a-prioi centroid 1', color='#006EB4') 
ax.scatter(centroid_2[0], centroid_2[1], s=50, label='a-prioi centroid 2', color='#0B2027')

ax.axvline(c='grey', lw=0.5)
ax.axhline(c='grey', lw=0.5)
    
ax.legend()
ax.set_xlim([-8,25])
ax.set_ylim([-1,25])
plt.show()

#fig.savefig('correlated_dataset.png', bbox_inches='tight', transparent=True)
```
