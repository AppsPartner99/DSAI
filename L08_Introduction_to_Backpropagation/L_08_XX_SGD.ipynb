```python
from lecture_utils import Training, alphaSmoothing
from matplotlib import pyplot as plt
```

```python
epochs = 30
training_loss_256 = Training(batchSize=256, epochs=epochs, learningRate= 0.01).train(maxIterations=10000)
training_loss_1 = Training(batchSize=1, epochs=epochs, learningRate= 0.01).train(maxIterations=10000)
training_loss_1024 = Training(batchSize=1024, epochs=epochs, learningRate= 0.01).train(maxIterations=10000)

```

```python
plt.plot(training_loss_256[0], label=r"$B = 256$")
plt.plot(training_loss_1[0], label=r"$B = 1$")
```

```python
plt.plot(alphaSmoothing(training_loss_256[1],0.99),'r',alpha=0.2)
plt.plot(alphaSmoothing(training_loss_1[1],0.3),'k',alpha=0.2)
plt.plot(alphaSmoothing(training_loss_1024[1],0.3),'g',alpha=0.2)
plt.plot(alphaSmoothing(training_loss_256[1],0.1),'r',label=r"$L_{B=256,\alpha=0.1}$")
plt.plot(alphaSmoothing(training_loss_1[1],0.01),'k',label=r"$L_{B=1,\alpha=0.01}$")
plt.plot(alphaSmoothing(training_loss_1024[1],0.1),'g',label=r"$L_{B=1024,\alpha=0.1}$")
plt.legend()
```
